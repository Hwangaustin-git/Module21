## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop 'EIN' and 'NAME' columns as they are not useful for prediction
application_df = application_df.drop(columns=['EIN', 'NAME'])

# Group rare occurrences in 'APPLICATION_TYPE' into "Other"
application_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_counts[application_counts < 500].index)

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app, "Other")

# Group rare occurrences in 'CLASSIFICATION' into "Other"
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_to_replace = list(classification_counts[classification_counts < 1000].index)

for cls in classification_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Convert categorical variables into numeric using get_dummies
application_df = pd.get_dummies(application_df)
```

```python

# Define features (X) and target (y)
X = application_df.drop(columns=['IS_SUCCESSFUL']).values
y = application_df['IS_SUCCESSFUL'].values

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the neural network model
nn = Sequential()

# Input layer & First hidden layer with more neurons
nn.add(Dense(units=100, activation='relu', input_dim=X_train.shape[1]))

# Second hidden layer with increased neurons
nn.add(Dense(units=50, activation='relu'))

# Third hidden layer for further complexity
nn.add(Dense(units=25, activation='relu'))

# Output layer
nn.add(Dense(units=1, activation='sigmoid'))

# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display model summary
nn.summary()
```

```python
# Train the model with more epochs
nn.fit(X_train_scaled, y_train, epochs=150, batch_size=32)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization

# Define the improved neural network model
nn_optimized = Sequential()

# Input layer & First hidden layer with more neurons and Swish activation
nn_optimized.add(Dense(units=256, activation='swish', input_dim=X_train.shape[1]))
nn_optimized.add(BatchNormalization())  # Normalize activations

# Second hidden layer with ELU activation for better non-linearity
nn_optimized.add(Dense(units=128, activation='elu'))
nn_optimized.add(BatchNormalization())

# Third hidden layer with more neurons
nn_optimized.add(Dense(units=64, activation='swish'))
nn_optimized.add(BatchNormalization())

# Fourth hidden layer with ELU activation
nn_optimized.add(Dense(units=32, activation='elu'))
nn_optimized.add(BatchNormalization())

# Add dropout layer to prevent overfitting
nn_optimized.add(Dropout(0.3))  # Drop 30% of neurons

# Output layer with sigmoid activation
nn_optimized.add(Dense(units=1, activation='sigmoid'))

# Compile the optimized model with lower learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)
nn_optimized.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Display model structure
nn_optimized.summary()
```

```python
# Train the optimized model with more epochs and smaller batch size
nn_optimized.fit(X_train_scaled, y_train, epochs=300, batch_size=32)
```

```python
# Evaluate the optimized model on the test set
model_loss, model_accuracy = nn_optimized.evaluate(X_test_scaled, y_test, verbose=2)

# Print results
print(f"Test Loss: {model_loss:.4f}")
print(f"Test Accuracy: {model_accuracy:.4f}")
```
